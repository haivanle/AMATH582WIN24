```python
# We will be working with the IRIS dataset (part of sample datasets in scikitlearn)
# We will cluster flowers types with KMeans and then use the clustering for classification

import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
y_names = iris.target_names

# print(X[40:60])
# print(y[40:60])
# print(y_names)
```

```python
from sklearn.model_selection import train_test_split

# use train_test_split to split the data randomly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

```

```python
# KNN Classifier applied to classify the flowers 
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)



KMCL = KMeans(n_clusters=3)
KMCL.fit(X_train, y_train)

y_pred = KMCL.fit_predict(X_train)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('Original Features Data (Training Set)')
plt.show()
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_pred)
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('KMeans Clustered Features Data (Training Set)')
plt.show()
```
